```python
import pytesseract
from PIL import Image
from gtts import gTTS
from IPython.display import Audio

```


# Step 1: Load image


```python
img = Image.open("quote.jpg")  # replace with your filename
```


# Step 2: OCR


```python
text = pytesseract.image_to_string(img)
print("Extracted Text:\n", text)

```

# Step 3: Text-to-Speech


```python
tts = gTTS(text)
tts.save("output.mp3")

```

# Step 4: Play audio


```python
Audio("output.mp3")
```
